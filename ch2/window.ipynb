```python
w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
    w.activate() # 창 활성화

if not w.isMaximized:
    w.maximize()

p.sleep(1)
# if not w.isMinimized:
#     w.minimize() # 최소화
w.restore() # 원래 크기로
```

```python
import pyperclip

w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
     w.activate()

p.write("write")
p.write("pyautogui", interval=0.25)
# p.write("안녕하세요") 한글 작성 안됨

pyperclip.copy("안녕하세요")
p.hotkey("ctrl","v")
```

```python
# 함수를 만들어서 한국어, 영어 상관없이 쓸수 있도록 만든거
def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
     w.activate()

word_write("write")
word_write("반값습니다")
```
